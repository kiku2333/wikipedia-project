```python
#!/usr/bin/python3

import dask
import dask.dataframe as ddf
import pandas as pd
import numpy as np
import pickle
```

```python
path = '../../../../intermediate-result-new-select-editors/warring-camps/words-added-removed-info-based-on-revert-graph-csv/'
```

```python
import glob
files = glob.glob(path + "*.csv")
```

```python
df = pd.DataFrame()
```

```python
for f in files:
    csv = pd.read_csv(f,sep='\t',encoding='utf-16')
    df = df.append(csv)
```

```python
with open('../../all-data-article-clustering/articles-100-threshold.pkl' ,'rb') as f:
    selected_articles = pickle.load(f) 
```

```python
df = df[df['page.title'].isin(selected_articles)]
```

```python
df.to_parquet('result/all-data-revert-user-index-info-with-words-added-removed')
```

```python
df.to_csv('result/all-data-revert-user-index-info-with-words-added-removed.csv',sep = '\t', encoding = 'utf-16')
```
