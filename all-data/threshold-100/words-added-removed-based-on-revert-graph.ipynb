```python
import dask
import dask.dataframe as ddf
import pandas as pd
import numpy as np
from string import punctuation
import collections
import re
import mwparserfromhell
from nltk.corpus import stopwords
import codecs
import os
from glob import glob
import pickle
import ast
```

```python
path = '../../../../intermediate-result-new-select-editors/warring-camps/words-added-removed-info-based-on-revert-graph-csv/'
```

```python
all_files = glob(path + '*')
```

```python
res = pd.DataFrame()
for i in all_files:
    df = pd.read_csv(i,sep = '\t', encoding = 'utf-16')
    res = res.append(df)
```

```python
with open('../../all-data-article-clustering/articles-100-threshold.pkl' ,'rb') as f:
    selected_articles = pickle.load(f) 
```

```python
res = res[res['page.title'].isin(selected_articles)]
```

```python
res = res.reset_index().drop(columns={'Unnamed: 0','index'})
```

```python
user_group = pd.read_csv('result/all-data-editor-cluster-result.csv', sep="\t",encoding='utf-16').set_index('contributor.username')
```

```python
group_revert_info = res.merge(user_group,left_on='user1',right_index=True)[['rev1','user1','added_words','removed_words','group','page.title']]
```

```python
def test(x):
    return x[8:-1]
```

```python
group_revert_info['added_words'] = group_revert_info['added_words'].map(test)
```

```python
group_revert_info['removed_words'] = group_revert_info['removed_words'].map(test)
```

```python
def convert_to_dict(a):
    if a == '':
        return collections.Counter()
    return collections.Counter(ast.literal_eval(a))
```

```python
group_revert_info['added_words'] = group_revert_info['added_words'].map(convert_to_dict)
```

```python
group_revert_info['removed_words'] = group_revert_info['removed_words'].map(convert_to_dict)
```

```python
fin = group_revert_info.groupby(['group','page.title']).agg({'added_words':'sum','removed_words':'sum'})
```

```python
fin['add.removed'] = fin['added_words'] - fin['removed_words']

fin['removed.add'] = fin['removed_words'] - fin['added_words']
```

```python
def to_negative(x):
    for k,v in x.items():
        x[k] = -v
    return x
```

```python
fin['removed.add'] = fin['removed.add'].apply(to_negative)
```

```python
def find_diff(x):
    a = x['add.removed']
    b = x['removed.add']
    return collections.Counter({x: a.get(x, 0) + b.get(x, 0) for x in set(a).union(b)})
```

```python
fin['diff'] = fin.apply(find_diff,axis=1)
```

```python
fin = fin[['added_words','removed_words','diff']]
```

```python
def sort_and_select_top_100(a):
    d= {}
    for k, v in sorted(a.items(), key=lambda x: abs(x[1]), reverse=True):
        d[k] = v
    return list(d.items())[:100]
```

```python
fin['diff'] = fin['diff'].apply(sort_and_select_top_100)
```

```python
def sort_counter(c):
    return c.most_common()[:100]
```

```python
fin['added_words'] = fin['added_words'].apply(sort_counter)
```

```python
fin['removed_words'] = fin['removed_words'].apply(sort_counter)
```

```python
fin.to_csv('result/added-removed-words-per-group-per-article-based-on-revert.csv',sep='\t',encoding='utf-16')
```
